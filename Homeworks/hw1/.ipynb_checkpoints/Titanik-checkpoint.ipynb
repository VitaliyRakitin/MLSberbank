Посмотрим, как вообще фичи связаны друг с другом.

```python
def plot_two_features_scatter(x_i, x_j, y):
    # Set colors and plot scatter
    colors = []
    for cls in y:
        if cls == 1.0:
            colors.append('r')
        if cls == 0.0:
            colors.append('g')
    plt.scatter(x_j, x_i, c=colors,s = 4, linewidths=0)
    return

    
def plot_feature_histogram(x_i, y):
    # Compute positive and negative histograms
    cls_0 = []
    cls_1 = []
    for i,cl in enumerate(y):
        if cl == 0.0:
            cls_1.append(x_i[i])
        else:
            cls_0.append(x_i[i])
    # Plot stacked barplots
    plt.hist((cls_1, cls_0), color=("r", "g"))
    return


def plot_dataset(x, y, features):
    # Tune your plot if necessary
    # your code here
    plt.figure(figsize=(20,20))
    
    for i, feature_i in enumerate(features):
        for j, feature_j in enumerate(features):

            # Tune your plot if necessary (for example set labels)
            size = (len(features), len(features))
            plot = plt.subplot2grid(size, (i, j))
            plot.set_xticklabels([])
            plot.set_yticklabels([])
            plot.xaxis.set_label_position("top")
            
            # Do actual plotting
            if i != j:
                plot_two_features_scatter(x[:, i], x[:, j], y)            
            else:
                plot_feature_histogram(x[:, i], y)
            
            if i == 0:
                plt.xlabel(feature_j)

            if j == 0:
                plt.ylabel(feature_i)
    
    plt.show()
```

Так как в данных присутствуют Nan, избавимся от них, а после построим нормальные графики.

#### Задание 3.2
Напишите функцию, которая будет превращать колонку с категориальными признаками в колонку с числовыми признаками:

(подсказка: будет работать быстрее, если работать не с pd.DataFrame, а перевести в np.array())

```python
categorical_columns = [c for c in data[goodFeatures].columns if data[c].dtype.name == 'object']
numerical_columns   = [c for c in data[goodFeatures].columns if data[c].dtype.name != 'object']
print (categorical_columns)
print (numerical_columns)
```

```python
def categorical_to_numeric(data, column):
    """
    column: колонка значений типа pd.DataFrame
    функция возвращает np.array() с новыми числовыми значениями этой колонки
    """    
    col = np.array(data[column])
    unique = data[column].unique()
    return np.array([np.where(unique == i) for i in col]).reshape(col.shape[0])
```

Переведем колонки и категориальными признаками в числовые с помощью только что написанной Вами функции:

***Замечание*** Считаю, что несовсем правильно безвозвратно затирать данные. Лучше генерировать новые features.

```python
data1= data[goodFeatures]
for column in data1.columns:
    if data1[column].dtypes == object:
        col = column + "_Fix"
        data[col] = categorical_to_numeric(data,column)
        goodFeatures.remove(column)
        goodFeatures.append(col)
```

#### Задание*

Для тех, кто на прошлом занятии игрался с knn.ipynb, предлагается применить knn к полученным данным, а потом получить ответы на данные из test_titakin.csv и загрузить в соревнование: https://www.kaggle.com/c/titanic. Какой результат получился и при каких k?

Нормируем данные. Каждый признак приведем к нулевому среднему и единичному среднеквадратичному отклонению.

```python
X_bad = data[goodFeatures].values 
```

```python
X = (X_bad - X_bad.mean()) / X_bad.std()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.15, random_state=37)
```

```python
target_length = float(len(Y_test)) 
quality_test = []
quality_train = []
for n in range(1,20):
    knn = KNeighborsClassifier(n_neighbors=n)
    cl = knn.fit(X_train, Y_train)
    pred_train = cl.predict(X_train)
    pred_test = cl.predict(X_test)
    quality_train.append(accuracy_score(Y_train,pred_train))
    quality_test.append(accuracy_score(Y_test,pred_test))
```

```python
plt.figure(figsize=(10,10))
plt.plot(quality_test, label = "train")
plt.plot(quality_train, label = "test")
plt.legend()
```

Качество на обучающей выборке при небольшом количестве соседей неплохое, чего не скажешь о тестовой.
